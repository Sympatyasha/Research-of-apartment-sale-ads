  * –î–æ–±–∞–≤—å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –∫ –∫–æ–¥—É, —á—Ç–æ–±—ã –ø–æ—è—Å–Ω–∏—Ç—å –µ–≥–æ –ª–æ–≥–∏–∫—É. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø–æ—á–µ–º—É –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è 1.5 * IQR –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≥—Ä–∞–Ω–∏—Ü –≤—ã–±—Ä–æ—Å–æ–≤.

**–í–æ—Ç –ø—Ä–∏–º–µ—Ä —Å —É—á–µ—Ç–æ–º —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–π:**   
    
    def anomaly_clear(data, column, remove_outliers=True):
        """
        –£–¥–∞–ª—è–µ—Ç –≤—ã–±—Ä–æ—Å—ã –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ column DataFrame data, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω (IQR).

        :param data: DataFrame —Å –¥–∞–Ω–Ω—ã–º–∏
        :param column: –ù–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–¥–∞–ª–∏—Ç—å –≤—ã–±—Ä–æ—Å—ã
        :param remove_outliers: –§–ª–∞–≥, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π, —Å–ª–µ–¥—É–µ—Ç –ª–∏ —É–¥–∞–ª—è—Ç—å –≤—ã–±—Ä–æ—Å—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é True)
        :return: –û—á–∏—â–µ–Ω–Ω—ã–π DataFrame (–µ—Å–ª–∏ remove_outliers=True) –∏–ª–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ —É–¥–∞–ª–µ–Ω–Ω—ã—Ö –≤—ã–±—Ä–æ—Å–æ–≤
        """
        print(f"–ò—Å—Ö–æ–¥–Ω—ã–π boxplot –¥–ª—è {column}")
        plt.figure(figsize=(6, 4))
        data.boxplot(column=column)
        plt.show()

        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1

        # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ª—É—á–∞—è, –∫–æ–≥–¥–∞ IQR —Ä–∞–≤–µ–Ω 0
        if IQR == 0:

            return data

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        print(f"–ì—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤ –¥–ª—è {column}: [{lower_bound}, {upper_bound}]")

        if remove_outliers:
            filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
            print(f"Boxplot –¥–ª—è {column} –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤")
            filtered_data.boxplot(column=column)
            plt.show()
            return filtered_data
        else:
            outliers_count = len(data[(data[column] < lower_bound) | (data[column] > upper_bound)])
            print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ {column}: {outliers_count}")
            return f"–í—ã–±—Ä–æ—Å—ã –Ω–µ —É–¥–∞–ª–µ–Ω—ã. –ò—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {outliers_count}"

```python
# –°–Ω–∞—á–∞–ª–∞ —Å–æ–∑–¥–∞–µ–º —Ç–∞–±–ª–∏—Ü—É —Å –º–µ–¥–∏–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
median_ceiling_height_by_location = data.groupby('locality_name')['ceiling_height'].median().reset_index(name='ceiling_height_median')

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å —Ç–∞–±–ª–∏—Ü–µ–π –º–µ–¥–∏–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
data_filled_by_location_corrected = pd.merge(data, median_ceiling_height_by_location, on='locality_name', how='left')

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ ceiling_height, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ –ª–æ–∫–∞—Ü–∏–∏
data_filled_by_location_corrected['ceiling_height'].fillna(data_filled_by_location_corrected['ceiling_height_median'], inplace=True)

# –ü–æ–≤—Ç–æ—Ä—è–µ–º –ø—Ä–æ—Ü–µ–¥—É—Ä—É –¥–ª—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏ –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –¥–æ —Ü–µ–Ω—Ç—Ä–∞
# –°–æ–∑–¥–∞–µ–º –≥—Ä—É–ø–ø—ã –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –¥–æ —Ü–µ–Ω—Ç—Ä–∞
data['cityCenters_nearest_group'] = pd.cut(data['cityCenters_nearest'], bins=10)
median_ceiling_height_by_distance = data.groupby('cityCenters_nearest_group')['ceiling_height'].median().reset_index(name='ceiling_height_median_by_distance')

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å —Ç–∞–±–ª–∏—Ü–µ–π –º–µ–¥–∏–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é
data_filled_by_distance_corrected = pd.merge(data, median_ceiling_height_by_distance, on='cityCenters_nearest_group', how='left')

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ ceiling_height, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –¥–æ —Ü–µ–Ω—Ç—Ä–∞
data_filled_by_distance_corrected['ceiling_height'].fillna(data_filled_by_distance_corrected['ceiling_height_median_by_distance'], inplace=True)

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è: —Å–∫–æ–ª—å–∫–æ –æ—Å—Ç–∞–ª–æ—Å—å –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ—Å–ª–µ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø–æ–¥—Ö–æ–¥–æ–≤
missing_after_location_corrected = data_filled_by_location_corrected['ceiling_height'].isna().sum()
missing_after_distance_corrected = data_filled_by_distance_corrected['ceiling_height'].isna().sum()

print("–°–∫–æ–ª—å–∫–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ—Å–ª–µ –ª–æ–∫–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –º–µ—Ç–æ–¥–∞ ", missing_after_location_corrected)
print("–°–∫–æ–ª—å–∫–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ—Å–ª–µ –º–µ—Ç–æ–¥–∞ –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –¥–æ —Ü–µ–Ω—Ç—Ä–∞ ", missing_after_distance_corrected)
```

```python
def floortype(floors):
    if floors['floor'] == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if floors['floor'] == floors['floors_total']:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–¥—Ä—É–≥–æ–π'
def floortypedigit(floors):
    if floors['floor'] == 1:
        return 1
    if floors['floor'] == floors['floors_total']:
        return 3
    return 2
data['floor_type'] = data.apply(floortype, axis=1)
data['floor_type_digit'] = data.apply(floortypedigit, axis=1)
data['floor_type'].head(20)
#data['floor_type_digit'].head(20)
```

    """
    –£–¥–∞–ª—è–µ—Ç –≤—ã–±—Ä–æ—Å—ã –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ column DataFrame data, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω (IQR).

def anomaly_clear(data, column, max_outliers_percentage=5):
    """
    –£–¥–∞–ª—è–µ—Ç –≤—ã–±—Ä–æ—Å—ã –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ column DataFrame data, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω (IQR),

    :param max_outliers_percentage: –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø—Ä–æ—Ü–µ–Ω—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –∫–∞–∫ –≤—ã–±—Ä–æ—Å—ã

    if IQR == 0:

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    if outliers_percentage > max_outliers_percentage:

        # –£–¥–∞–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç—É —á–∞—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä–∞—è –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç max_outliers_percentage
        num_outliers_to_remove = int(len(data) * max_outliers_percentage / 100)
        filtered_data = data.drop(outliers.nsmallest(num_outliers_to_remove, column).index)
    else:
        filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return filtered_data
